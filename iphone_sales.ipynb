```python
sorted_data = df.sort_values(by='Star Rating', ascending=False)

# Selecting the top 10 highest-rated iPhones
top_10_rated = sorted_data.head(10)

# Plotting a bar chart using Matplotlib
plt.figure(figsize=(10, 6))
bars=plt.bar(top_10_rated['Product Name'], top_10_rated['Star Rating'], color='skyblue',align="center")
# Add values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval,round(yval , 2), ha='center', va='bottom')
plt.xlabel('Names')
plt.ylabel('Star Rating')
plt.xticks(rotation=90)
plt.title('Top 10 Highest-Rated iPhones')
plt.show()
```

From above diagram we can conclude that "Apple iphone 11 Pro MAx(Midnight Green , 64 GB)" is the rated highest among the 10 highest ratings
means that more people are fond of this model
 

```python
bars2=plt.bar(top_10_rated['Product Name'], top_10_rated['Number Of Ratings'], color='skyblue',align="center")
for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval,round(yval , 2), ha='center', va='bottom')

plt.xlabel('Names')
plt.ylabel('Number Of Ratings')
plt.xticks(rotation=90)
plt.title('number of ratings of the highest-rated iPhone')
plt.show()
```

From Above graph I conclude that no. of rating for this model "APPLE Iphone 8 plus(Gold,64 GB)" is the highest among the above top 10 highest rated Iphone

```python
# Number of reviews of the highest-rated iPhone
bars2=plt.bar(top_10_rated['Product Name'], top_10_rated['Number Of Reviews'], color='skyblue',align="center")
for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval,round(yval , 2), ha='center', va='bottom')

plt.xlabel('Names')
plt.ylabel('Number Of Reviews')
plt.xticks(rotation=90)
plt.title('number of reviews of the highest-rated iPhone')
plt.show()
```

FRom the above graph I conclude taht "APPLE iPhone 8 Plus (Gold, 64 GB)" is also leading in the highest number of reviews on Flipkart among the highest-rated iPhones in India.

```python
x=df["Number Of Ratings"]
y=df["Sale Price"]
plt.scatter(x,y,color='skyblue',alpha=1)
plt.xlabel('Sale Price')
plt.ylabel('Number Of Ratings')
plt.title("Relationship between Sale Price and Number of Ratings of iPhones")
plt.show()


```

From above graph I conclude that iPhones with lower sale prices are sold more in India.

```python
x=df["Discount Percentage"]
y=df["Number Of Ratings"]
# plt.figure(figsize=(20,15))
plt.xlabel('Discount %')
plt.ylabel('Number Of Ratings')
plt.scatter(x,y,color='r',alpha=0.4)
plt.title("Relationship between Discount Percentage and Number of Ratings of iPhones", fontsize=10)
plt.show()


```

From above graph I conclude that the iphones with less number of ratings are getting discount only in range 5% to 15% and iphones with higher number of ratings are given discount in range of 15% to 25%.

Overall concluding after looking these graphs , APPLE iPhone 8 Plus (Gold, 64 GB) was the most appreciated iPhone in India . IPhones with lower sale prices are sold more in India.
